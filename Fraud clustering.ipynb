```python
# import the libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.preprocessing import StandardScaler
%matplotlib inline
```

```python
# read our dataset
df = pd.read_csv('creditcard.csv')
df.head()
```

```python
df.info()
```

```python
# df.value_count(df.Class.unique())
df["Class"].value_counts()
```

```python
s_scaler = StandardScaler()
df[["Amount"]] = s_scaler.fit_transform(df[["Amount"]])
```

```python
df["Time_Diff"] = df.Time.diff()
df.head()
```

```python
# resample the data

df_Fraud = df[df.Class == 1]
# df_nonFraud_indesxes = (np.random.choice(df[df.Class == 0 ].index , len(df_Fraud)))
df_nonFraud = df[df.Class == 0 ].sample(len(df_Fraud))
new_df = pd.concat([df_Fraud , df_nonFraud])
new_df = new_df.sample(frac =1)
new_df.head(10)
```

```python
x = new_df.drop(columns = ["Class" , "Time"])
y = pd.DataFrame(new_df.Class)
```

```python
# PCA Feature Selection (1)

from sklearn.decomposition import PCA
pca = PCA (n_components= 0.95, random_state=1 )
pca_new_x = pca.fit_transform(x)
pca_new_x.shape
```

```python
# K-Mean Clustering

from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 2, random_state = 0)
kmeans_predictions = kmeans.fit_predict(pca_new_x)
```

```python
Clustering_K = list(kmeans.labels_)
y ["Clustering_K"] = Clustering_K
y.head()
```

```python
from sklearn.metrics import confusion_matrix


CM_K = confusion_matrix(y.Class, y.Clustering_K)
print (CM_K)
```

```python
from sklearn.metrics import accuracy_score
AccScore_K = accuracy_score(y.Class, y.Clustering_K, normalize=False)

print('Accuracy Score is : ', AccScore_K)
```

```python
from sklearn.metrics import f1_score
F1Score_K = f1_score(y.Class, y.Clustering_K, average='micro')
print('F1 Score is : ', F1Score_K)
```

```python
# Check Clustering Data

if F1Score_K < 0.45:
    y["Clustering_K"] = y["Clustering_K"].replace({0:1 , 1:0})
    CM_K = confusion_matrix(y.Class, y.Clustering_K)
    print (CM_K)
    AccScore_K = accuracy_score(y.Class, y.Clustering_K, normalize=False)
    print('Accuracy Score is : ', AccScore_K)
    F1Score_K = f1_score(y.Class, y.Clustering_K, average='micro')
    print('F1 Score is : ', F1Score_K)
```

```python
# Fitting Hierarchical Clustering to the dataset
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 2)
y_hc = hc.fit_predict(x)
Clustering_HC = list(y_hc)
y ["Clustering_HC"] = Clustering_HC
y.head()
```

```python
from sklearn.metrics import confusion_matrix
CM_hc = confusion_matrix(y.Class, y.Clustering_HC)
print (CM_hc)
```

```python
from sklearn.metrics import accuracy_score
AccScore_hc = accuracy_score(y.Class, y.Clustering_HC, normalize=False)

print('Accuracy Score is : ', AccScore_hc)
```

```python
from sklearn.metrics import f1_score
F1Score_hc = f1_score(y.Class, y.Clustering_HC, average='micro')
print('F1 Score is : ', F1Score_hc)
```
