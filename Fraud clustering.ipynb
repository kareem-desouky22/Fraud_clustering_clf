```python
# import the libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.preprocessing import StandardScaler
%matplotlib inline
```

```python
# read our dataset
df = pd.read_csv('creditcard.csv')
df.head()
```

```python
df.info()
```

```python
# df.value_count(df.Class.unique())
df["Class"].value_counts()
```

```python
s_scaler = StandardScaler()
df[["Amount" , "Time"]] = s_scaler.fit_transform(df[["Amount" , "Time"]])
```

```python
# resample the data

df_Fraud = df[df.Class == 1]
# df_nonFraud_indesxes = (np.random.choice(df[df.Class == 0 ].index , len(df_Fraud)))
df_nonFraud = df[df.Class == 0 ].sample(len(df_Fraud))
new_df = pd.concat([df_Fraud , df_nonFraud])
new_df = new_df.sample(frac =1)
new_df.head()
```

```python
x = new_df.drop(columns = "Class")
y = pd.DataFrame(new_df.Class)
```

```python
# PCA Feature Selection

from sklearn.decomposition import PCA
pca = PCA (n_components= 0.95, random_state=1 )
pca_new_x = pca.fit_transform(x)
np.sum(pca.explained_variance_ratio_)
```

```python
# K-Mean Clustering

from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 2, random_state = 0)
kmeans_predictions = kmeans.fit_predict(pca_new_x)
```

```python
kmeans.cluster_centers_
```

```python
v = list(kmeans.labels_)
y ["clustering"] = v
y
```

```python
from sklearn.metrics import confusion_matrix


CM = confusion_matrix(y.Class, y.clustering)
print (CM)
```

```python
from sklearn.metrics import accuracy_score
AccScore = accuracy_score(y.Class, y.clustering, normalize=False)

print('Accuracy Score is : ', AccScore)
```

```python
from sklearn.metrics import f1_score
F1Score = f1_score(y.Class, y.clustering, average='micro')
print('F1 Score is : ', F1Score)
```
